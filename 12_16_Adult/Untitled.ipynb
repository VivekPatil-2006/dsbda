```python
# Import pandas
# Used for data analysis
import pandas as pd


# ------------------------------------------------
# Load Adult Dataset
# header=None means no column names
# ------------------------------------------------
data = pd.read_csv(
    "adult.csv",
    header=None
)


# ------------------------------------------------
# Add column names
# ------------------------------------------------
data.columns = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education_num",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital_gain",
    "capital_loss",
    "hours_per_week",
    "native_country",
    "income"
]


# ------------------------------------------------
# Show first 5 rows
# ------------------------------------------------
print(data.head())



# =================================================
# s. CREATE SUBSETS
# =================================================

# Get India data
india_data = data[
    data["native_country"] == "India"
]


# Get Male data
male_data = data[
    data["sex"] == "Male"
]


# Get White race data
white_data = data[
    data["race"] == "White"
]


# Show India data
print(india_data.head())

# Show Male data
print(male_data.head())

# Show White race data
print(white_data.head())



# =================================================
# t. MERGE TWO SUBSETS
# =================================================

# concat() joins datasets
merged_data = pd.concat([
    india_data,
    male_data
])


# Show merged data
print(merged_data.head())



# =================================================
# u. SORT DATA
# =================================================

# sort_values() sorts data
#
# ascending=False means descending order
sorted_data = data.sort_values(
    "age",
    ascending=False
)


# Show sorted data
print(sorted_data.head())



# =================================================
# v. TRANSPOSE DATA
# =================================================

# .T converts rows into columns
transpose_data = data.head().T


# Show transposed data
print(transpose_data)



# =================================================
# w. MELT DATA
# =================================================

# melt() converts wide format into long format
melt_data = pd.melt(

    # Dataset
    data,

    # Fixed column
    id_vars=["sex"],

    # Columns to melt
    value_vars=[
        "age",
        "hours_per_week"
    ]
)


# Show melt data
print(melt_data.head())



# =================================================
# x. CAST TO WIDE FORMAT
# =================================================

# pivot_table() converts long format into wide format
wide_data = melt_data.pivot_table(

    # Row names
    index="sex",

    # Column names
    columns="variable",

    # Values
    values="value",

    # mean() calculates average
    aggfunc="mean"
)


# Show wide data
print(wide_data)
```
