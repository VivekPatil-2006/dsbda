```python
import pandas as pd

# Load dataset without headers
data = pd.read_csv("adult.csv", header=None)

# Add column names manually
data.columns = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education_num",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital_gain",
    "capital_loss",
    "hours_per_week",
    "native_country",
    "income"
]

# Show dataset
print(data.head())


# s. Create subsets for Country, Sex, Race

india_data = data[data["native_country"] == "India"]

male_data = data[data["sex"] == "Male"]

white_data = data[data["race"] == "White"]

print(india_data.head())
print(male_data.head())
print(white_data.head())


# t. Merge two subsets
merged_data = pd.concat([india_data, male_data])

print(merged_data.head())


# u. Sort Data using age
sorted_data = data.sort_values("age", ascending=False)

print(sorted_data.head())


# v. Transpose Data
transpose_data = data.head().T

print(transpose_data)


# w. Melt Data to long format
melt_data = pd.melt(
    data,
    id_vars=["sex"],
    value_vars=["age", "hours_per_week"]
)

print(melt_data.head())


# x. Cast Data to wide format
wide_data = melt_data.pivot_table(
    index="sex",
    columns="variable",
    values="value",
    aggfunc="mean"
)

print(wide_data)
```
