```python
# Import BeautifulSoup
# Used to read HTML file
from bs4 import BeautifulSoup

# Import pandas
# Used to create table and CSV file
import pandas as pd


# ------------------------------------------------
# Open index.html file
# "r" = read mode
# ------------------------------------------------
file = open("index.html", "r")


# ------------------------------------------------
# Read HTML code
# ------------------------------------------------
soup = BeautifulSoup(file, "html.parser")


# ------------------------------------------------
# Find all review blocks
#
# "div" = HTML tag
# class_="review" = review class
# ------------------------------------------------
reviews = soup.find_all("div", class_="review")


# Empty list
data = []


# ------------------------------------------------
# Read reviews one by one
# ------------------------------------------------
for review in reviews:

    # Customer name
    customer = review.find("h3", class_="name").text.strip()

    # Review comment
    comment = review.find("p", class_="comment").text.strip()

    # Rating
    rating = review.find("span", class_="rating").text.strip()

    # Tag
    tag = review.find("span", class_="tag").text.strip()


    # Store data
    data.append([
        customer,
        comment,
        rating,
        tag
    ])


# ------------------------------------------------
# Create table
# ------------------------------------------------
df = pd.DataFrame(
    data,
    columns=[
        "Customer Name",
        "Review",
        "Rating",
        "Tag"
    ]
)


# ------------------------------------------------
# Show data
# ------------------------------------------------
print(df)


# ------------------------------------------------
# Save data into CSV file
#
# index=False removes extra numbering
# ------------------------------------------------
df.to_csv("pict_reviews.csv", index=False)


# Message
print("Reviews saved successfully")
```
