```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL
url = "https://webscraper.io/test-sites/e-commerce/static"

# Send request
response = requests.get(url)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find products/reviews
reviews = soup.find_all("div", class_="thumbnail")

# Empty list
data = []

# Loop through reviews
for review in reviews:

    # Customer Name (Demo Name)
    customer_name = "PICT Student"

    # Comment / Review
    comment = review.find("p", class_="description")

    if comment:
        comment = comment.text.strip()
    else:
        comment = "No Review"

    # Rating
    stars = review.find_all("span", class_="glyphicon-star")

    rating = len(stars)

    # Tags
    if rating >= 4:
        tag = "Excellent"
    elif rating >= 3:
        tag = "Good"
    else:
        tag = "Average"

    # Store data
    data.append([
        customer_name,
        comment,
        rating,
        tag
    ])

# Create DataFrame
df = pd.DataFrame(
    data,
    columns=[
        "Customer Name",
        "Review",
        "Rating",
        "Tag"
    ]
)

# Display data
print(df)

# Save to CSV
df.to_csv("pict_reviews.csv", index=False)

print("Reviews saved to pict_reviews.csv")
```
