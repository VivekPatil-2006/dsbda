```python
# Import requests library
# Used to send request to website
import requests

# Import BeautifulSoup class
# Used to read and extract HTML data
from bs4 import BeautifulSoup

# Import pandas library
# Used to create table and CSV file
import pandas as pd


# ------------------------------------------------
# Website URL
# ------------------------------------------------
url = "https://webscraper.io/test-sites/e-commerce/allinone"


# ------------------------------------------------
# Send request to website
# requests.get() gets webpage data
# ------------------------------------------------
response = requests.get(url)


# ------------------------------------------------
# Read HTML code of webpage
# response.text = complete HTML code
# "html.parser" is used to parse/read HTML
# ------------------------------------------------
soup = BeautifulSoup(response.text, "html.parser")


# ------------------------------------------------
# Find all product cards
# find_all() finds multiple matching tags
#
# "div" = HTML tag
# class_="thumbnail" = product container class
# ------------------------------------------------
products = soup.find_all("div", class_="thumbnail")


# Empty list
# Used to store all product data
data = []


# ------------------------------------------------
# Read products one by one
# ------------------------------------------------
for product in products:

    # ------------------------------------------------
    # Product Name
    #
    # find() finds first matching tag
    #
    # "a" = anchor tag
    # class_="title" = product title class
    #
    # .text = get text only
    # .strip() = remove extra spaces
    # ------------------------------------------------
    name = product.find("a", class_="title").text.strip()


    # ------------------------------------------------
    # Product Price
    #
    # "h4" = heading tag
    # class_="price" = price class
    # ------------------------------------------------
    price = product.find("h4", class_="price").text.strip()


    # ------------------------------------------------
    # Product Description
    #
    # "p" = paragraph tag
    # class_="description" = description class
    # ------------------------------------------------
    comment = product.find("p", class_="description").text.strip()


    # ------------------------------------------------
    # Product Rating
    #
    # find_all() gets all star icons
    # Each star = one rating
    #
    # len() counts total stars
    # ------------------------------------------------
    rating = product.find_all("span", class_="ws-icon-star")

    rating = len(rating)


    # ------------------------------------------------
    # Review Count
    #
    # class_="review-count" stores review text
    # ------------------------------------------------
    reviews = product.find("p", class_="review-count")


    # Check if reviews exist
    if reviews:

        # Get review text
        reviews = reviews.text.strip()

    else:

        # If reviews not found
        reviews = "No Reviews"


    # ------------------------------------------------
    # Store all product data into list
    # ------------------------------------------------
    data.append([
        name,
        price,
        comment,
        rating,
        reviews
    ])


# ------------------------------------------------
# Create DataFrame (table)
#
# columns=[] sets column names
# ------------------------------------------------
df = pd.DataFrame(
    data,
    columns=[
        "Name",
        "Price",
        "Comment",
        "Rating",
        "Reviews"
    ]
)


# ------------------------------------------------
# Display first 5 rows
# head() shows top rows
# ------------------------------------------------
print(df.head())


# ------------------------------------------------
# Save table into CSV file
#
# index=False removes extra numbering column
# ------------------------------------------------
df.to_csv("ecommerce_reviews.csv", index=False)


# Message
print("Data saved successfully")
```
