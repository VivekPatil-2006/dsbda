```python
# Import pandas
# Used for data analysis
import pandas as pd


# ------------------------------------------------
# Load Hepatitis Dataset
# ------------------------------------------------
data = pd.read_csv("hepatitis.csv")
data.columns = [
    "Class",
    "AGE",
    "SEX",
    "STEROID",
    "ANTIVIRALS",
    "FATIGUE",
    "MALAISE",
    "ANOREXIA",
    "LIVER_BIG",
    "LIVER_FIRM",
    "SPLEEN_PALPABLE",
    "SPIDERS",
    "ASCITES",
    "VARICES",
    "BILIRUBIN",
    "ALK_PHOSPHATE",
    "SGOT",
    "ALBUMIN",
    "PROTIME",
    "HISTOLOGY"
]

# ------------------------------------------------
# Show first 5 rows
# ------------------------------------------------
print(data.head())



# =================================================
# a. CREATE SUBSETS
# =================================================

# Get Male data
male_data = data[
    data["SEX"] == "male"
]


# Get Female data
female_data = data[
    data["SEX"] == "female"
]


# Show Male data
print(male_data.head())

# Show Female data
print(female_data.head())



# =================================================
# b. MERGE TWO SUBSETS
# =================================================

# concat() joins datasets
merged_data = pd.concat([
    male_data,
    female_data
])


# Show merged data
print(merged_data.head())



# =================================================
# c. SORT DATA
# =================================================

# sort_values() sorts data
#
# Sort using AGE, SGOT and PROTIME
sorted_data = data.sort_values([
    "AGE",
    "SGOT",
    "PROTIME"
])


# Show sorted data
print(sorted_data.head())



# =================================================
# d. TRANSPOSE DATA
# =================================================

# .T converts rows into columns
transpose_data = data.head().T


# Show transposed data
print(transpose_data)



# =================================================
# e. MELT DATA
# =================================================

# melt() converts wide format into long format
melt_data = pd.melt(

    # Dataset
    data,

    # Fixed column
    id_vars=["SEX"],

    # Columns to melt
    value_vars=[
        "AGE",
        "SGOT"
    ]
)


# Show melt data
print(melt_data.head())



# =================================================
# f. CAST TO WIDE FORMAT
# =================================================

# pivot_table() converts long format into wide format
wide_data = melt_data.pivot_table(

    # Row names
    index="SEX",

    # Column names
    columns="variable",

    # Values
    values="value",

    # mean() calculates average
    aggfunc="mean"
)


# Show wide data
print(wide_data)
```
