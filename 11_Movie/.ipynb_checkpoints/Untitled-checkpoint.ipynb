```python
import pandas as pd

# Load dataset
data = pd.read_csv("movies_metadata.csv", low_memory=False)

# Show first rows
print(data.head())


# m. Create subsets for different languages
english_movies = data[data["original_language"] == "en"]

french_movies = data[data["original_language"] == "fr"]

print(english_movies.head())
print(french_movies.head())


# n. Merge two subsets
merged_data = pd.concat([english_movies, french_movies])

print(merged_data.head())


# o. Sort Data using customer ratings
sorted_data = data.sort_values("vote_average", ascending=False)

print(sorted_data.head())


# p. Transpose Data
transpose_data = data.head().T

print(transpose_data)


# q. Melt Data to Long Format
melt_data = pd.melt(
    data,
    id_vars=["original_language"],
    value_vars=["vote_average", "vote_count"]
)

print(melt_data.head())


# r. Cast Data to Wide Format
wide_data = melt_data.pivot_table(
    index="original_language",
    columns="variable",
    values="value",
    aggfunc="mean"
)

print(wide_data)
```
