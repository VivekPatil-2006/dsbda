```python
import pandas as pd

# Load Forest Fires Dataset
data = pd.read_csv("forestfires.csv")

print(data.head())


# ------------------------------------------------
# a. CREATE DATA SUBSETS
# ------------------------------------------------

# Create classes using area affected

not_affected = data[data["area"] == 0]

partially_affected = data[
    (data["area"] > 0) & (data["area"] <= 50)
]

mostly_affected = data[data["area"] > 50]

print("Not Affected")
print(not_affected.head())

print("Partially Affected")
print(partially_affected.head())

print("Mostly Affected")
print(mostly_affected.head())


# ------------------------------------------------
# b. MERGE TWO SUBSETS
# ------------------------------------------------

merged_data = pd.concat([
    partially_affected,
    mostly_affected
])

print("Merged Data")
print(merged_data.head())


# ------------------------------------------------
# c. SORT DATA
# ------------------------------------------------

sorted_data = data.sort_values(
    ["temp", "wind", "area"],
    ascending=False
)

print("Sorted Data")
print(sorted_data.head())


# ------------------------------------------------
# d. TRANSPOSE DATA
# ------------------------------------------------

transpose_data = data.head().T

print("Transpose Data")
print(transpose_data)


# ------------------------------------------------
# e. MELT DATA TO LONG FORMAT
# ------------------------------------------------

melt_data = pd.melt(
    data,
    id_vars=["month"],
    value_vars=["temp", "wind", "area"]
)

print("Melt Data")
print(melt_data.head())


# ------------------------------------------------
# f. CAST DATA TO WIDE FORMAT
# ------------------------------------------------

wide_data = melt_data.pivot_table(
    index="month",
    columns="variable",
    values="value",
    aggfunc="mean"
)

print("Wide Format Data")
print(wide_data)
```
