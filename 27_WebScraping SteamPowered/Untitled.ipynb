```python
# Import requests
# Used to get data from website/API
import requests

# Import pandas
# Used to create table and CSV file
import pandas as pd


# ------------------------------------------------
# Steam Review API URL
# ------------------------------------------------
url = "https://store.steampowered.com/appreviews/730?json=1"


# ------------------------------------------------
# Send request to website
# ------------------------------------------------
response = requests.get(url)


# ------------------------------------------------
# Convert response into JSON format
# ------------------------------------------------
data = response.json()


# ------------------------------------------------
# Get all reviews
# ------------------------------------------------
reviews = data["reviews"]


# Empty list to store review data
review_list = []


# ------------------------------------------------
# Read each review one by one
# ------------------------------------------------
for review in reviews:

    # Customer Steam ID
    customer = review["author"]["steamid"]

    # Review comment
    comment = review["review"]

    # Rating
    # voted_up = True or False
    if review["voted_up"]:
        rating = "Positive"
    else:
        rating = "Negative"

    # Helpful votes
    votes = review["votes_up"]


    # Store all data in list
    review_list.append([
        customer,
        comment,
        rating,
        votes
    ])


# ------------------------------------------------
# Create DataFrame (Table)
# ------------------------------------------------
df = pd.DataFrame(
    review_list,
    columns=[
        "Customer",
        "Comment",
        "Rating",
        "Helpful Votes"
    ]
)


# ------------------------------------------------
# Display Reviews
# ------------------------------------------------
print(df)


# ------------------------------------------------
# Save reviews into CSV file
# ------------------------------------------------
df.to_csv("steam_reviews.csv", index=False)


# Message
print("Reviews saved successfully")
```
