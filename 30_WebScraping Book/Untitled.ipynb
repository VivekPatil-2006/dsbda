```python
# Import requests
# Used to get website data
import requests

# Import BeautifulSoup
# Used to read HTML code
from bs4 import BeautifulSoup

# Import pandas
# Used to create table and CSV file
import pandas as pd


# ------------------------------------------------
# Website URL
# ------------------------------------------------
url = "https://books.toscrape.com/"


# ------------------------------------------------
# Get website data
# requests.get() sends request to website
# ------------------------------------------------
response = requests.get(url)


# ------------------------------------------------
# Read HTML code
#
# response.text = webpage HTML
# "html.parser" reads HTML
# ------------------------------------------------
soup = BeautifulSoup(response.text, "html.parser")


# ------------------------------------------------
# Find all books
#
# "article" = HTML tag
# class_="product_pod" = book container
# ------------------------------------------------
books = soup.find_all("article", class_="product_pod")


# Empty list
# Used to store book data
data = []


# ------------------------------------------------
# Read books one by one
# ------------------------------------------------
for book in books:

    # ------------------------------------------------
    # Book Name
    #
    # ["title"] gets title attribute
    # ------------------------------------------------
    name = book.h3.a["title"]


    # ------------------------------------------------
    # Rating
    #
    # Rating is stored in class name
    # Example:
    # class = "star-rating Three"
    #
    # [1] gets "Three"
    # ------------------------------------------------
    rating = book.p["class"][1]


    # ------------------------------------------------
    # Book Price
    #
    # .text gets text only
    # ------------------------------------------------
    price = book.find("p", class_="price_color").text


    # Demo Comment Tag
    comment = "Book Review"


    # Demo Customer Name
    customer = "Book Customer"


    # ------------------------------------------------
    # Store data into list
    # ------------------------------------------------
    data.append([
        customer,
        name,
        rating,
        comment,
        price
    ])


# ------------------------------------------------
# Create table
#
# columns=[] sets column names
# ------------------------------------------------
df = pd.DataFrame(
    data,
    columns=[
        "Customer Name",
        "Book Name",
        "Rating",
        "Comment Tag",
        "Price"
    ]
)


# ------------------------------------------------
# Show first 5 rows
# head() displays top rows
# ------------------------------------------------
print(df.head())


# ------------------------------------------------
# Save data into CSV file
#
# index=False removes numbering column
# ------------------------------------------------
df.to_csv("book_reviews.csv", index=False)


# Message
print("Data saved successfully")
```
