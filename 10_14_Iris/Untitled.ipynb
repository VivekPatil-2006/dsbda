```python
# Import pandas
# Used for data analysis
import pandas as pd


# ------------------------------------------------
# Load Iris Dataset
# header=None means dataset has no column names
# ------------------------------------------------
data = pd.read_csv("Iris.csv", header=None)


# ------------------------------------------------
# Add column names
# ------------------------------------------------
data.columns = [
    "SepalLength",
    "SepalWidth",
    "PetalLength",
    "PetalWidth",
    "Species"
]


# ------------------------------------------------
# Show first 5 rows
# ------------------------------------------------
print(data.head())



# =================================================
# g. CREATE SUBSETS
# =================================================

# Get only Setosa flowers
setosa = data[data["Species"] == "Iris-setosa"]

# Get only Versicolor flowers
versicolor = data[data["Species"] == "Iris-versicolor"]


# Show Setosa data
print(setosa.head())

# Show Versicolor data
print(versicolor.head())



# =================================================
# h. MERGE TWO SUBSETS
# =================================================

# concat() joins two datasets
merged_data = pd.concat([
    setosa,
    versicolor
])


# Show merged data
print(merged_data.head())



# =================================================
# i. SORT DATA
# =================================================

# sort_values() sorts data
# by PetalLength column
sorted_data = data.sort_values(
    "PetalLength"
)


# Show sorted data
print(sorted_data.head())



# =================================================
# j. TRANSPOSE DATA
# =================================================

# .T converts rows into columns
transpose_data = data.head().T


# Show transposed data
print(transpose_data)



# =================================================
# k. MELT DATA
# =================================================

# melt() converts wide format into long format
melt_data = pd.melt(

    # Dataset
    data,

    # Fixed column
    id_vars=["Species"],

    # Columns to melt
    value_vars=[
        "SepalLength",
        "PetalLength"
    ]
)


# Show melt data
print(melt_data.head())



# =================================================
# l. CAST TO WIDE FORMAT
# =================================================

# pivot_table() converts long format into wide format
wide_data = melt_data.pivot_table(

    # Row names
    index="Species",

    # Column names
    columns="variable",

    # Values
    values="value"
)


# Show wide data
print(wide_data)
```
