```python
# Import pandas library
# Used to read CSV file
import pandas as pd

# Import matplotlib library
# Used to create graphs
import matplotlib.pyplot as plt


# ------------------------------------------------
# Load Dataset
# header=None means dataset has no column names
# ------------------------------------------------
data = pd.read_csv("Cleavland.csv", header=None)



# =================================================
# a. HISTOGRAM
# =================================================

# Histogram is used to show distribution of data

# Column 0 = Age
plt.hist(data[0])

# Title of graph
plt.title("Histogram of Age")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Frequency")

# Show graph
plt.show()



# =================================================
# b. DOT PLOT
# =================================================

# Dot plot is used to show individual values

# Column 4 = Cholesterol
# 'o' means display points as dots
plt.plot(data[4], 'o')

# Title of graph
plt.title("Dot Plot of Cholesterol")

# X-axis name
plt.xlabel("Patient")

# Y-axis name
plt.ylabel("Cholesterol")

# Show graph
plt.show()



# =================================================
# c. BAR PLOT
# =================================================

# Bar plot is used to compare values

# Column 13 = Target (Heart Disease)
# value_counts() counts repeated values
# kind='bar' creates bar graph
data[13].value_counts().plot(kind='bar')

# Title of graph
plt.title("Bar Plot of Heart Disease")

# X-axis name
plt.xlabel("Target")

# Y-axis name
plt.ylabel("Count")

# Show graph
plt.show()



# =================================================
# d. LINE CHART
# =================================================

# Line chart is used to show variation or trend

# Column 7 = Maximum Heart Rate
plt.plot(data[7])

# Title of graph
plt.title("Line Chart of Heart Rate")

# X-axis name
plt.xlabel("Patient")

# Y-axis name
plt.ylabel("Heart Rate")

# Show graph
plt.show()



# =================================================
# e. BOX PLOT + HISTOGRAM + SCATTER PLOT
# =================================================

# Column 4 = Cholesterol

# Histogram
plt.hist(data[4])

# Scatter Plot
plt.scatter(range(len(data[4])), data[4])

# Box Plot
plt.boxplot(data[4])

# Title
plt.title("Box Plot with Histogram and Scatter Plot")

# Show graph
plt.show()
```
