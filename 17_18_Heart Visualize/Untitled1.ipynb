```python
# Import pandas
# Used to read CSV file
import pandas as pd

# Import matplotlib
# Used to create graphs
import matplotlib.pyplot as plt


# ------------------------------------------------
# Load Dataset
# header=None means no column names in dataset
# ------------------------------------------------
data = pd.read_csv("Cleavland.csv", header=None)



# =================================================
# a. HISTOGRAM
# =================================================

# Histogram shows data distribution

# Column 0 = Age
plt.hist(data[0])

# Graph title
plt.title("Histogram of Age")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Frequency")

# Show graph
plt.show()



# =================================================
# b. PIE CHART
# =================================================

# Pie chart shows percentage

# Column 13 = Heart Disease
data[13].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

# Graph title
plt.title("Pie Chart of Heart Disease")

# Remove ylabel
plt.ylabel("")

# Show graph
plt.show()



# =================================================
# c. BOX PLOT
# =================================================

# Box plot shows spread and outliers

# Column 4 = Cholesterol
plt.boxplot(data[4])

# Graph title
plt.title("Box Plot of Cholesterol")

# Y-axis name
plt.ylabel("Cholesterol")

# Show graph
plt.show()



# =================================================
# d. SCATTER PLOT
# =================================================

# Scatter plot shows relation between 2 values

# Column 0 = Age
# Column 4 = Cholesterol
plt.scatter(data[0], data[4])

# Graph title
plt.title("Scatter Plot of Age vs Cholesterol")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Cholesterol")

# Show graph
plt.show()



# =================================================
# e. SCATTER PLOT + BOX PLOT
# =================================================

# Scatter plot
plt.scatter(data[0], data[4])

# Box plot
plt.boxplot(data[4], positions=[80])

# Graph title
plt.title("Scatter Plot with Box Plot")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Cholesterol")

# Show graph
plt.show()
```
