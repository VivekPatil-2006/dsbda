```python
import pandas as pd
```

```python
data=pd.read_csv("dataset_Facebook.csv",sep=";")
data.head();
```

```python
photo_posts = data[data["Type"] == "Photo"]

status_posts = data[data["Type"] == "Status"]

print(photo_posts.head())

print(status_posts.head())
```

```python
merged_data = pd.concat([photo_posts, status_posts])

print(merged_data.head())
```

```python
sorted_data = data.sort_values("Page total likes")

print(sorted_data.head())
```

```python
transpose_data = data.head().T

print(transpose_data)
```

```python
melt_data = pd.melt(
    data,
    id_vars=["Type"],
    value_vars=["like", "share"]
)

print(melt_data.head())
```

```python
wide_data = melt_data.pivot_table(
    index="Type",
    columns="variable",
    values="value"
)

print(wide_data)
```
