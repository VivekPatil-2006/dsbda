```python
# Import pandas
# Used to read CSV file
import pandas as pd

# Import matplotlib
# Used to create graphs
import matplotlib.pyplot as plt


# ------------------------------------------------
# Load Dataset
# header=None means dataset has no column names
# ------------------------------------------------
data = pd.read_csv("adult.csv", header=None)



# =================================================
# f. HISTOGRAM
# =================================================

# Histogram shows distribution of data

# Column 0 = Age
plt.hist(data[0])

# Graph title
plt.title("Histogram of Age")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Frequency")

# Show graph
plt.show()



# =================================================
# g. DOT PLOT
# =================================================

# Dot plot shows individual values

# Column 12 = Hours Per Week
# 'o' means dots
plt.plot(data[12], 'o')

# Graph title
plt.title("Dot Plot of Working Hours")

# X-axis name
plt.xlabel("Person")

# Y-axis name
plt.ylabel("Hours Per Week")

# Show graph
plt.show()



# =================================================
# h. BAR PLOT
# =================================================

# Bar plot compares values

# Column 14 = Income
data[14].value_counts().plot(kind='bar')

# Graph title
plt.title("Bar Plot of Income Category")

# X-axis name
plt.xlabel("Income")

# Y-axis name
plt.ylabel("Count")

# Show graph
plt.show()



# =================================================
# i. LINE CHART
# =================================================

# Line chart shows variation or trend

# Column 10 = Capital Gain
plt.plot(data[10])

# Graph title
plt.title("Line Chart of Capital Gain")

# X-axis name
plt.xlabel("Person")

# Y-axis name
plt.ylabel("Capital Gain")

# Show graph
plt.show()



# =================================================
# j. HISTOGRAM + SCATTER PLOT + BOX PLOT
# =================================================

# Column 0 = Age


# Histogram
plt.hist(data[0])

# Scatter Plot
plt.scatter(range(len(data[0])), data[0])

# Box Plot
plt.boxplot(data[0])

# Graph title
plt.title("Histogram + Scatter Plot + Box Plot")

# Show graph
plt.show()
```
