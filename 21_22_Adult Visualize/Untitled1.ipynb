```python
# Import pandas
# Used to read CSV file
import pandas as pd

# Import matplotlib
# Used to create graphs
import matplotlib.pyplot as plt


# ------------------------------------------------
# Load Dataset
# header=None means no column names
# ------------------------------------------------
data = pd.read_csv("adult.csv", header=None)



# =================================================
# a. HISTOGRAM
# =================================================

# Histogram shows data distribution

# Column 0 = Age
plt.hist(data[0])

# Graph title
plt.title("Histogram of Age")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Frequency")

# Show graph
plt.show()



# =================================================
# b. PIE CHART
# =================================================

# Pie chart shows percentage

# Column 14 = Income
data[14].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

# Graph title
plt.title("Pie Chart of Income")

# Remove ylabel
plt.ylabel("")

# Show graph
plt.show()



# =================================================
# c. BOX PLOT
# =================================================

# Box plot shows spread and outliers

# Column 12 = Hours Per Week
plt.boxplot(data[12])

# Graph title
plt.title("Box Plot of Working Hours")

# Y-axis name
plt.ylabel("Hours Per Week")

# Show graph
plt.show()



# =================================================
# d. SCATTER PLOT
# =================================================

# Scatter plot shows relation between 2 values

# Column 0 = Age
# Column 12 = Hours Per Week
plt.scatter(data[0], data[12])

# Graph title
plt.title("Scatter Plot of Age vs Working Hours")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Hours Per Week")

# Show graph
plt.show()



# =================================================
# e. ADD BOXPLOT TO SCATTERPLOT
# =================================================

# Scatter plot
plt.scatter(data[0], data[12])

# Box plot
plt.boxplot(data[12], positions=[90])

# Graph title
plt.title("Scatter Plot with Box Plot")

# X-axis name
plt.xlabel("Age")

# Y-axis name
plt.ylabel("Hours Per Week")

# Show graph
plt.show()
```
